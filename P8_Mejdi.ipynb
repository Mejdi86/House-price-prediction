### Pipelines pour l'imputation de valeurs nulles, la mise à l'échelle des colonnes numériques et l'encodage des colonnes catégorielles

```python
from sklearn.impute import SimpleImputer
# imputer for numeric columns
num_imputer = SimpleImputer(missing_values=np.nan, strategy='median')
# imputer for categorical columns
cat_imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
```

```python
from sklearn.preprocessing import StandardScaler
# normalized scaling for numerical columns
num_scaler = StandardScaler()
```

```python
from sklearn.preprocessing import OneHotEncoder
# encoder for categorical columns
cat_encoder = OneHotEncoder(handle_unknown = 'ignore')
```

```python
from sklearn.pipeline import Pipeline

# pipeline for numerical columns
num_pipeline = Pipeline([
    ('imputer', num_imputer),
    ('std_scaler', num_scaler)
])

# pipeline for categorical columns
cat_pipeline = Pipeline([
    ('imputer', cat_imputer),
     ('cat', cat_encoder)
])
```

```python
from sklearn.compose import ColumnTransformer

# due to dropping of some columns seperating column names again
num_columns = X_train.select_dtypes(np.number).columns
cat_columns = X_train.select_dtypes(object).columns
    
full_pipeline = ColumnTransformer([
        ("num", num_pipeline, num_columns),
        ("cat", cat_pipeline, cat_columns)
    ])

X_train = full_pipeline.fit_transform(X_train)
# resultant X_train is compressed sparse matrix, not original dataframe
```
